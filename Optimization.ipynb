```
import pandas as pd
def Optimise(txt):
  x = txt.split()
  name = []
  min = [] 
  max = []
  avg = []
  uni = []
  request = []
  limit = []
  remark = []
  for i in range(int(len(x)/7)):
    s=i*7
    #if(validate(x[s],x[s+1],x[s+3],x[s+5],float(x[s+5])/0.7,float(x[s+5])/0.65,x[s+6])):
      #return
    name.append(x[s])
    min.append(x[s+1])
    max.append(x[s+3])
    avg.append(x[s+5])
    uni.append(x[s+6])
    #print(x[s+6])
    if x[s+6] == 'GiB':
      request.append(round(float(x[s+5])/0.7,1))
      limitX = round(float(x[s+5])/0.65,1)
      limit.append(limitX)
    elif x[s+6] == 'MiB':
      request.append(int(round(float(x[s+5])/0.7)))
      limitX = int(round(float(x[s+5])/0.65))
      limit.append(limitX)
    else:
      request.append(0)
      limit.append(0)

    if (0.8*limitX < float(x[s+3])):
      remark.append('max is ' + str(round(float(x[s+3])/limitX*100))+ '% of max ' + str(round(limitX/0.8,2)))
    else:
      remark.append('Ok')

  data = {'Name':name,'Min':min,'Max':max,'Avg':avg,'Request':request,'Limit':limit,'Uni':uni,'Remark':remark}
  df = pd.DataFrame(data)
  return df
  
def validate(conf,txt):
  x = txt.split()
  name = []
  min = [] 
  max = []
  avg = []
  uni = []
  request = []
  limit = []
  remark = []
  for i in range(int(len(x)/7)):
    s=i*7
    #if(validate(x[s],x[s+1],x[s+3],x[s+5],float(x[s+5])/0.7,float(x[s+5])/0.65,x[s+6])):
      #return
    name.append(x[s])
    min.append(x[s+1])
    max.append(x[s+3])
    avg.append(x[s+5])
    uni.append(x[s+6])
    

  data = {'Name':name,'Min':min,'Max':max,'Avg':avg}
  df = pd.DataFrame(data)
  request_perc = []
  limit_perc = []

  for i in range(df.shape[0]):
    if(float(opt.iloc[i]['Request'])):
      request_perc.append(float(df.iloc[i]['Avg'])/float(opt.iloc[i]['Request']))
    else:
      request_perc.append(0)
    if(float(opt.iloc[i]['Limit'])):
      limit_perc.append(float(df.iloc[i]['Avg'])/float(opt.iloc[i]['Limit']))
    else:
      limit_perc.append(0)
    #limit_perc.append(df.iloc[i]['Avg']/opt.iloc[i]['Limit'])
  data_new = {'Name':name,'Request':request_perc,'Limit':limit_perc}
  dfX = pd.DataFrame(data_new) 
  return dfX 
```
